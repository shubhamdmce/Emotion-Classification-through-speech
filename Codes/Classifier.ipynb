```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
from sklearn import preprocessing
```

```python
df_cry=pd.read_csv('adult_cry2.csv')
df_laugh=pd.read_csv('adult_laugh2.csv')
```

```python
df_cry.head()
```

```python
df_laugh.head()
```

```python
frames = [df_cry, df_laugh]
df= pd.concat(frames)
```

```python
Y=df['emotion']
X=df.drop(['emotion'],axis=1)
col_names=list(X.columns.values)
```

```python
X1 = X.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(X1)
X = pd.DataFrame(x_scaled)
X.columns=col_names
```

```python
X.head()
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25)
```

```python
from sklearn import svm
clf = svm.SVC()
clf.fit(X_train, Y_train) 
```

```python
Y_pred=clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test, Y_pred)*100
```

```python
from sklearn.model_selection import cross_val_score
clf = svm.SVC()
scores=cross_val_score(clf, X, Y, cv=5, n_jobs=-1)    #10 fold Cross Validation
scores
```
